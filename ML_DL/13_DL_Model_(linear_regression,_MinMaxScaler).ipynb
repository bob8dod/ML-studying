```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split #for spliting data
from sklearn.preprocessing import MinMaxScaler #For normalization _ preprocessing

import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
#Setting random seed
seed=3
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
#Setting Data
df = pd.read_csv('./dataset/housing.csv', delim_whitespace=True, header=None)
df.head()
```

```python
#Data Preprocessing
data = df.values

s=MinMaxScaler() #for normalization
s.fit(data)
data_s = s.transform(data)

X = data_s[:,0:-1].astype(float)
Y = data_s[:,[-1]].astype(float)
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state = seed) #split data for test
```

```python
print(data_s)
```

```python
#Make Model
model = Sequential()
model.add(Dense(30,input_dim=13,activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(1))
```

```python
#Set Model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
#Operate Model
history = model.fit(x_train,y_train,epochs=200, batch_size=10)
```

```python
#Test the Model
Y_prediction = model.predict(x_test).flatten() #Using flatten to make 1-D array
for i in range(10):
    label = y_test[i]
    prediction = Y_prediction[i]
    print("실제가격: %.4f, 예상가격: %.4f" %(label, prediction))
```
