```python
#Using Package
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder #For transfromimg string to integer 
from keras.utils import np_utils #For One-Hot Encoding

import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Set random seed
seed=3
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
#Set Data
df = pd.read_csv('./dataset/iris.csv', names=["sepal_length","sepa_width","petal_length","petal_width","species"])
df.head(5)
```

```python
#Showing Relation Map
sns.pairplot(df, hue='species')
plt.show()
```

```python
#Data preprocessing
data = df.values
X = data[:,0:4].astype('float64')
Y_obj = data[:,4] # type : string

# transform string to integer and One-Hot Encoding
e = LabelEncoder() 
e.fit(Y_obj)
Y = e.transform(Y_obj)
Y = np_utils.to_categorical(Y) #One-Hot Encoding
```

```python
#Make Model
model = Sequential()
model.add(Dense(16,input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax')) #using 'softmax!' _ for multinominal
```

```python
#Setting Model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) #multinominal classification
```

```python
#Operating Model
model.fit(X,Y,epochs=100,batch_size=10)
print('Accuracy: %.4f' %(model.evaluate(X,Y)[1]))
```
