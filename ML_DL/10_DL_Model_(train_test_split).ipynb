```python
#Using Package
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
#Set random seed
seed = 3
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
#Set Data
df = pd.read_csv('./dataset/sonar.csv', header=None)
df.head()
```

```python
#Checking the Label _ Y_obj
df[[60]].describe()
```

```python
#Data Preprocessing
data = df.values
X = data[:,0:60].astype(float)
Y_obj = data[:,60]

#Transforming the Y (str->int)
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

#Split the Data(train, test)
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=seed)
```

```python
#Make Model
model = Sequential()
model.add(Dense(24,input_dim=60,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
#Setting Model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=130, batch_size=5)
# model.save('my_model.h5') #Save the Model on Computer
```

```python
# del model                              #Delete Model in here for Test
# model = load_model('my_model.h5')      #Load Saved Model
print('Accuracy: %.4f' %(model.evaluate(x_test, y_test)[1])) #Evaluate the Model
```
