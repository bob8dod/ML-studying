```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold

import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
seed = 3
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
df = pd.read_csv('./dataset/sonar.csv', header=None)
df.head()
```

```python
df[[60]].describe()
```

```python
#Data Preprocessing
data = df.values
X = data[:,0:60].astype(float)
Y_obj = data[:,60]

#Data transforming
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
#KFold _ data split
n_fold = 10
skf=StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)
```

```python
#Model Making,Setting,Operating
accuracy=[] #for saving all accuracy
for train,test in skf.split(X,Y): # 10 iterations
    model=Sequential()
    model.add(Dense(24,input_dim=60,activation='relu'))
    model.add(Dense(10,activation='relu'))
    model.add(Dense(1,activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.fit(X[train],Y[train],epochs=30, batch_size=10) #Fitting by train_set
    accuracy_k = "%.4f" %(model.evaluate(X[test], Y[test])[1]) #Evaluating by test_set
    accuracy.append(accuracy_k) #Saving as "string" not real number
```

```python
#Showing result
for i,value in enumerate(accuracy):
    print('[%02d]Flod:'%(i+1),value)
```
